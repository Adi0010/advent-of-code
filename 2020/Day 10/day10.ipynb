```python
s = set([int(x) for x in open('input.txt').read().split('\n')])
s.add(max(list(s)) + 3)

cur = 0
d1 = 0
d3 = 1

while len(s) > 1:
  print(len(s), cur, d1, d3)
  for x in s:
    if x - cur == 1:
      d1 += 1
      s.remove(x)
      cur = x
      break
    elif x - cur == 2:
      s.remove(x)
      cur = x
      break
    elif x - cur == 3:
      s.remove(x)
      cur = x
      d3 += 1
      break

print(d1, d3)
print(d1 * d3)
```

```python
v = [int(x) for x in open('input.txt').read().split('\n')]
v.append(0)
v.sort()
n = len(v)
dp = [0] * n
dp[0] = 1

for i in range(1, n):
  for j in range(i):
    if (v[i] - v[j] <= 3):
      dp[i] += dp[j]
      print(v[i], v[j], dp[i], dp[j])

print(dp[n - 1])
```
